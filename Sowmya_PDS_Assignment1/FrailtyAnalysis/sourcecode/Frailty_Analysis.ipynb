```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# # The graph below shows the distribution of Age for the given data

```python
ax=sns.distplot(df['Age'], kde = False, color ='blue', bins = 30)
fig=ax.get_figure()
fig.savefig("../Results/Age_Distibution.png")
```

```python
ax=sns.distplot(df['GripStrength'], kde = False, color ='green', bins = 30)
fig=ax.get_figure()
fig.savefig("../Results/GripStrength_Distibution.png")
```

```python
ax = df.plot.scatter(x='Age',y='GripStrength')
fig=ax.get_figure()
fig.savefig("../Results/GripStrength_Age_Relation.png")
```
